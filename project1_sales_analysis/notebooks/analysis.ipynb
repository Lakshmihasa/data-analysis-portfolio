```python
# Project 1: Business Sales Analysis (Superstore)

## Day 1 – Data Loading & Initial Exploration
```

```python
# Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df[['Order Date', 'Ship Date']].head()
```

```python
# Extract year and month
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.month_name()

df[['Order Date', 'Order Year', 'Order Month', 'Order Month Name']].head()
```

```python
monthly_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
monthly_sales.head()
```

```python
plt.figure(figsize=(4,4))
plt.plot(monthly_sales['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Time")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
#Objective:
#Analyze sales and profit performance across product categories and regions 
#to identify high-performing segments and business opportunities.
```

```python
# Category-wise Sales and Profit Summary
# Group data by Category and calculate total Sales and Profit
category_summary = (
    df.groupby("Category")[["Sales", "Profit"]]
      .sum()
      .reset_index()
)

category_summary

```

```python
# =========================
# Visualization: Sales by Category
# =========================

plt.figure(figsize=(5, 3))

sns.barplot(
    data=category_summary,
    x="Category",
    y="Sales"
)

plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.tight_layout()

plt.show()
```

```python
# =========================
# Visualization: Profit by Category
# =========================

plt.figure(figsize=(5, 3))

sns.barplot(
    data=category_summary,
    x="Category",
    y="Profit"
)

plt.title("Total Profit by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Profit")
plt.tight_layout()

plt.show()
```

```python
## Category Insights

# The Technology category generates the highest overall sales.
# Furniture shows strong revenue but comparatively lower profitability.
# Office Supplies contribute steady profit margins.
# Sales volume does not always directly translate to high profit.
```

```python
# =========================
# Region-wise Sales and Profit Summary
# =========================

region_summary = (
    df.groupby("Region")[["Sales", "Profit"]]
      .sum()
      .reset_index()
)

region_summary
```

```python
# =========================
# Visualization: Sales by Region
# =========================

plt.figure(figsize=(5, 3))

sns.barplot(
    data=region_summary,
    x="Region",
    y="Sales"
)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.tight_layout()

plt.show()
```

```python
## Regional Insights

# The West region leads in total sales performance.
# The South region shows comparatively lower sales.
# Regional differences highlight potential opportunities for targeted expansion strategies.
```

```python
# Day 4 – Discount vs Profit Analysis
# =========================

#Objective:
#Analyze the relationship between discount levels and profit 
#to understand how pricing strategies impact business performance.
```

```python
# Scatter Plot: Discount vs Profit
plt.figure(figsize=(5, 3))

sns.scatterplot(
    data=df,
    x="Discount",
    y="Profit"
)

plt.title("Relationship Between Discount and Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.tight_layout()

plt.savefig("../visuals/discount_vs_profit_scatter.png")

plt.show()
```

```python
# =========================
# Average Profit by Discount Level
# =========================

discount_profit_summary = (
    df.groupby("Discount")["Profit"]
      .mean()
      .reset_index()
)

discount_profit_summary
```

```python
# =========================
# Line Plot: Average Profit by Discount
# =========================

plt.figure(figsize=(5, 3))

sns.lineplot(
    data=discount_profit_summary,
    x="Discount",
    y="Profit",
    marker="o"
)

plt.title("Average Profit at Different Discount Levels")
plt.xlabel("Discount")
plt.ylabel("Average Profit")
plt.tight_layout()

plt.savefig("../visuals/avg_profit_by_discount.png")

plt.show()
```

```python
## Discount vs Profit Insights

# Profit declines as discount increases.
# High discount levels (e.g., 50% and above) frequently result in negative average profit.
# While discounts may increase order volume, they significantly reduce profitability.
# Businesses should reconsider deep discount strategies unless offset by volume gains.
```

```python
# =========================
# Filter orders where profit is negative
# =========================

loss_orders = df[df["Profit"] < 0]

loss_orders.head()
```

```python
# Count total loss-making orders
loss_count = loss_orders.shape[0]

print(f"Total loss-making orders: {loss_count}")
```

```python
# Calculate total loss amount
total_loss = loss_orders["Profit"].sum()

print(f"Total loss amount: {total_loss:.2f}")
```

```python
# =========================
# Profit by Sub-Category
# =========================

subcategory_profit = (
    df.groupby("Sub-Category")["Profit"]
      .sum()
      .reset_index()
      .sort_values(by="Profit")
)

subcategory_profit
```

```python
# =========================
# Bar Plot: Profit by Sub-Category
# =========================

plt.figure(figsize=(5, 3))

sns.barplot(
    data=subcategory_profit,
    x="Profit",
    y="Sub-Category"
)

plt.title("Profit by Sub-Category")
plt.xlabel("Total Profit")
plt.ylabel("Sub-Category")
plt.tight_layout()

plt.savefig("../visuals/profit_by_subcategory.png")

plt.show()
```

```python
# =========================
# Total Profit by Category
# =========================

category_profit_summary = (
    df.groupby("Category")["Profit"]
      .sum()
      .reset_index()
      .sort_values(by="Profit")
)

category_profit_summary
```

```python
## Loss Analysis Insights

# A significant number of orders result in negative profit.
# Certain sub-categories consistently generate losses.
# High-discount products are strongly associated with loss-making transactions.
# The Furniture category shows lower profitability compared to Technology.
# Targeted pricing and discount optimization strategies could improve margins.
```

```python
# =========================
# Customer-level Sales & Profit Summary
# =========================

customer_summary = (
    df.groupby("Customer Name")[["Sales", "Profit"]]
      .sum()
      .reset_index()
      .sort_values(by="Sales", ascending=False)
)

customer_summary.head()
```

```python
# =========================
# Top 10 Customers
# =========================

top_10_customers = customer_summary.head(10)

plt.figure(figsize=(5, 3))

sns.barplot(
    data=top_10_customers,
    x="Sales",
    y="Customer Name"
)

plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Name")
plt.tight_layout()

plt.savefig("../visuals/top_10_customers.png")

plt.show()
```

```python
# =========================
# Calculate contribution of top 20% customers
# =========================

total_revenue = customer_summary["Sales"].sum()

top_20_percent_count = int(0.2 * len(customer_summary))

top_20_revenue = customer_summary.head(top_20_percent_count)["Sales"].sum()

percentage_contribution = (top_20_revenue / total_revenue) * 100

print(f"Top 20% customers contribute {percentage_contribution:.2f}% of total revenue")
```

```python
## Customer Insights

# A small percentage of customers contribute a large portion of total revenue.
# Revenue concentration indicates potential dependency risk.
# High-value customers should be retained through loyalty strategies.
```

```python
# =========================
# Load Data into SQLite Database
# =========================

import sqlite3

# Create connection
conn = sqlite3.connect("../data/sales_database.db")

# Export dataframe to SQL table
df.to_sql("sales_data", conn, if_exists="replace", index=False)

print("Database created and data loaded successfully.")
```

```python
# =========================
# SQL Query: Total Sales by Category
# =========================

query = """
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM sales_data
GROUP BY Category
ORDER BY Total_Sales DESC;
"""

sql_result = pd.read_sql(query, conn)

sql_result
```

```python
import sqlite3
conn = sqlite3.connect("../data/sales_database.db")
```

```python
# =========================
# SQL: Top 5 Customers by Sales
# =========================

query_top_customers = """
SELECT "Customer Name",
       SUM(Sales) AS Total_Sales
FROM sales_data
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 5;
"""

top_customers_sql = pd.read_sql(query_top_customers, conn)

top_customers_sql
```

```python
# =========================
# SQL: Sub-Categories with Negative Profit
# =========================

query_loss_subcategories = """
SELECT "Sub-Category",
       SUM(Profit) AS Total_Profit
FROM sales_data
GROUP BY "Sub-Category"
HAVING Total_Profit < 0
ORDER BY Total_Profit;
"""

loss_subcategories_sql = pd.read_sql(query_loss_subcategories, conn)

loss_subcategories_sql
```

```python
# =========================
# SQL: Rank Customers by Sales
# =========================

query_rank_customers = """
SELECT
    "Customer Name",
    SUM(Sales) AS Total_Sales,
    RANK() OVER (ORDER BY SUM(Sales) DESC) AS Sales_Rank
FROM sales_data
GROUP BY "Customer Name";
"""

ranked_customers = pd.read_sql(query_rank_customers, conn)

ranked_customers.head()
```
